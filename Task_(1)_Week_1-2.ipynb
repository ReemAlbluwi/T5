## The dataset contains information about road traffic accidents with the following columns:

`Time`: Time of the accident

`Day_of_week`: Day of the week when the accident occurred

`Age_band_of_driver`: Age group of the driver

`Sex_of_driver`: Gender of the driver

`Educational_level`: Educational level of the driver

`Vehicle_driver_relation`: Relationship of the driver to the vehicle

`Driving_experience`: Driving experience of the driver

`Type_of_vehicle`: Type of vehicle involved in the accident

`Owner_of_vehicle`: Owner of the vehicle

`Service_year_of_vehicle`: Service years of the vehicle

`Area_accident_occured`: Area where the accident occurred

`Lanes_or_Medians`: Lanes or medians at the accident site

`Road_allignment`: Road alignment at the accident site

`Types_of_Junction`: Type of junction at the accident site

`Road_surface_type`: Type of road surface at the accident site

`Road_surface_conditions`: Road surface conditions at the accident site

`Light_conditions`: Light conditions at the time of the accident

`Weather_conditions`: Weather conditions at the time of the accident

`Type_of_collision`: Type of collision

`Number_of_vehicles_involved`: Number of vehicles involved in the accident

`Number_of_casualties`: Number of casualties in the accident

`Vehicle_movement`: Movement of the vehicle during the accident

`Casualty_class`: Class of casualty (driver, passenger, pedestrian)

`Sex_of_casualty`: Gender of the casualty

`Age_band_of_casualty`: Age group of the casualty

`Casualty_severity`: Severity of the casualty

`Work_of_casuality`: Occupation of the casualty

`Fitness_of_casuality`: Fitness of the casualty

`Pedestrian_movement`: Movement of the pedestrian

`Cause_of_accident`: Cause of the accident

`Accident_severity`: Severity of the accident

### Correct any inconsistent data entries.

```python
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce').dt.time
```

```python

df['Day_of_week'] = df['Day_of_week'].str.strip().str.lower()
df['Sex_of_driver'] = df['Sex_of_driver'].str.strip().str.lower()
df['Educational_level'] = df['Educational_level'].str.strip().str.lower()
df['Vehicle_driver_relation'] = df['Vehicle_driver_relation'].str.strip().str.lower()
df['Type_of_vehicle'] = df['Type_of_vehicle'].str.strip().str.lower()
df['Owner_of_vehicle'] = df['Owner_of_vehicle'].str.strip().str.lower()
df['Service_year_of_vehicle'] = df['Service_year_of_vehicle'].str.strip().str.lower()
df['Vehicle_movement'] = df['Vehicle_movement'].str.strip().str.lower()
df['Pedestrian_movement'] = df['Pedestrian_movement'].str.strip().str.lower()
df['Cause_of_accident'] = df['Cause_of_accident'].str.strip().str.lower()
df['Accident_severity'] = df['Accident_severity'].str.strip().str.lower()
```

```python
df['Type_of_vehicle'].fillna('NaN', inplace=True)
df['Service_year_of_vehicle'].fillna('NaN', inplace=True)
df['Casualty_class'].replace('na', pd.NA, inplace=True)
df['Sex_of_casualty'].replace('na', pd.NA, inplace=True)
df['Age_band_of_casualty'].replace('na', pd.NA, inplace=True)
df['Casualty_severity'].replace('na', pd.NA, inplace=True)
df['Work_of_casuality'].replace('na', pd.NA, inplace=True)
```

```python
df['Casualty_severity'] = df['Casualty_severity'].astype('Int32')
df['Casualty_class'].fillna('NaN', inplace=True)
df['Sex_of_casualty'].fillna('NaN', inplace=True)
df['Age_band_of_casualty'].fillna('NaN', inplace=True)
df['Work_of_casuality'].fillna('NaN', inplace=True)
```

**Findings Based on the Available Data:**

1. **Accident Peaks on Fridays:**
   - Fridays experience the highest number of accidents, with incidents exceeding 2000.

2. **Severity Distribution:**
   - Most accidents result in minor injuries, with fatalities accounting for less than 1.3% of cases.

3. **Age Group Impact:**
   - Drivers aged 18-30 are the most involved in accidents, indicating the need for targeted safety measures for this age group.

4. **Cause and Severity of Accidents:**
   - Failures in maintaining distance lead to accidents that generally have minor severity.

5. **Gender Impact:**
   - Male drivers are more frequently involved in accidents, but these accidents tend to be less severe.

6. **Education Level:**
   - Accidents are more common among individuals with lower educational levels.

7. **Weather Conditions:**
   - Accidents are more frequent under normal weather conditions, suggesting that weather is not a major factor in the severity of accidents.
